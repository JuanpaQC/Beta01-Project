```python
# function to convert the JavaScript object into an OpenCV image
def js_to_image(js_reply):
  """
  Params:
          js_reply: JavaScript object containing image from webcam
  Returns:
          img: OpenCV BGR image
  """
  # decode base64 image
  image_bytes = b64decode(js_reply.split(',')[1])
  # convert bytes to numpy array
  jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
  # decode numpy array into OpenCV BGR image
  img = cv2.imdecode(jpg_as_np, flags=1)

  return img

# function to convert OpenCV Rectangle bounding box image into base64 byte string to be overlayed on video stream
def bbox_to_bytes(bbox_array):
  """
  Params:
          bbox_array: Numpy array (pixels) containing rectangle to overlay on video stream.
  Returns:
        bytes: Base64 image byte string
  """
  # convert array into PIL image
  bbox_PIL = PIL.Image.fromarray(bbox_array, 'RGBA')
  iobuf = io.BytesIO()
  # format bbox into png for return
  bbox_PIL.save(iobuf, format='png')
  # format return string
  bbox_bytes = 'data:image/png;base64,{}'.format((str(b64encode(iobuf.getvalue()), 'utf-8')))

  return bbox_bytes

def video_frame(label, bbox):
  data = eval_js('stream_frame("{}", "{}")'.format(label, bbox))
  return data

# JavaScript to properly create our live video stream using our webcam as input
def video_stream():
  js = Javascript('''
    var video;
    var div = null;
    var stream;
    var captureCanvas;
    var imgElement;
    var labelElement;

    var pendingResolve = null;
    var shutdown = false;

    function removeDom() {
       stream.getVideoTracks()[0].stop();
       video.remove();
       div.remove();
       video = null;
       div = null;
       stream = null;
       imgElement = null;
       captureCanvas = null;
       labelElement = null;
    }

    function onAnimationFrame() {
      if (!shutdown) {
        window.requestAnimationFrame(onAnimationFrame);
      }
      if (pendingResolve) {
        var result = "";
        if (!shutdown) {
          captureCanvas.getContext('2d').drawImage(video, 0, 0, 800, 600);
          result = captureCanvas.toDataURL('image/jpeg', 0.8)
        }
        var lp = pendingResolve;
        pendingResolve = null;
        lp(result);
      }
    }

    async function createDom() {
      if (div !== null) {
        return stream;
      }

      div = document.createElement('div');
      div.style.border = '2px solid black';
      div.style.padding = '3px';
      div.style.width = '100%';
      div.style.maxWidth = '600px';
      document.body.appendChild(div);

      const modelOut = document.createElement('div');
      modelOut.innerHTML = "<span>Status:</span>";
      labelElement = document.createElement('span');
      labelElement.innerText = 'No data';
      labelElement.style.fontWeight = 'bold';
      modelOut.appendChild(labelElement);
      div.appendChild(modelOut);

      video = document.createElement('video');
      video.style.display = 'block';
      video.width = div.clientWidth - 6;
      video.setAttribute('playsinline', '');
      video.onclick = () => { shutdown = true; };
      stream = await navigator.mediaDevices.getUserMedia(
          {video: { facingMode: "environment"}});
      div.appendChild(video);

      imgElement = document.createElement('img');
      imgElement.style.position = 'absolute';
      imgElement.style.zIndex = 1;
      imgElement.onclick = () => { shutdown = true; };
      div.appendChild(imgElement);

      const instruction = document.createElement('div');
      instruction.innerHTML =
          '<span style="color: red; font-weight: bold;">' +
          'When finished, click here or on the video to stop this demo</span>';
      div.appendChild(instruction);
      instruction.onclick = () => { shutdown = true; };

      video.srcObject = stream;
      await video.play();

      captureCanvas = document.createElement('canvas');
      captureCanvas.width = 800; //video.videoWidth;
      captureCanvas.height = 600; //video.videoHeight;
      window.requestAnimationFrame(onAnimationFrame);

      return stream;
    }
    async function stream_frame(label, imgData) {
      if (shutdown) {
        removeDom();
        shutdown = false;
        return '';
      }

      var preCreate = Date.now();
      stream = await createDom();

      var preShow = Date.now();
      if (label != "") {
        labelElement.innerHTML = label;
      }

      if (imgData != "") {
        var videoRect = video.getClientRects()[0];
        imgElement.style.top = videoRect.top + "px";
        imgElement.style.left = videoRect.left + "px";
        imgElement.style.width = videoRect.width + "px";
        imgElement.style.height = videoRect.height + "px";
        imgElement.src = imgData;
      }

      var preCapture = Date.now();
      var result = await new Promise(function(resolve, reject) {
        pendingResolve = resolve;
      });
      shutdown = false;

      return {'create': preShow - preCreate,
              'show': preCapture - preShow,
              'capture': Date.now() - preCapture,
              'img': result};
    }
    ''')

  display(js)
```

```python
def calcular_edad(fecha_nacimiento):
    # Convierte la fecha de nacimiento en un objeto datetime
    fecha_nacimiento = datetime.strptime(fecha_nacimiento, '%Y-%m-%d')

    # Obtiene la fecha actual
    fecha_actual = datetime.now()

    # Calcula la diferencia en años entre la fecha actual y la fecha de nacimiento
    edad = fecha_actual.year - fecha_nacimiento.year

    # Ajusta la edad si la fecha actual aún no ha alcanzado el cumpleaños de la persona
    if fecha_actual.month < fecha_nacimiento.month or (fecha_actual.month == fecha_nacimiento.month and fecha_actual.day < fecha_nacimiento.day):
        edad -= 1

    return edad
```

```python
def Arbolbinary():

  #Calcula las fechas para sacar edades
  fecha = str(student_date_of_birth_input.value)
  ages = calcular_edad(fecha)

  class ArbolBinario:
      def __init__(self):
          self.root = None

      def insertar(self, estudiante):
          if not self.root:
              self.root = estudiante
          else:
              self._insertar_recursivo(estudiante, self.root)

      def _insertar_recursivo(self, estudiante, nodo_actual):
          if estudiante.age < nodo_actual.age:
              if nodo_actual.left is None:
                  nodo_actual.left = estudiante
              else:
                  self._insertar_recursivo(estudiante, nodo_actual.left)
          elif estudiante.age > nodo_actual.age:
              if nodo_actual.right is None:
                  nodo_actual.right = estudiante
              else:
                  self._insertar_recursivo(estudiante, nodo_actual.right)

      def en_orden(self, nodo_actual):
          if nodo_actual:
            self.en_orden(nodo_actual.left)
            #print(f"ID: {nodo_actual.id}, Name: {nodo_actual.name}, Age: {nodo_actual.age}")
            print(f'ID: {nodo_actual.id}','\n')
            print(f'Name: {nodo_actual.name}','\n')
            print(f'Age: {nodo_actual.age}','\n')
            print(f'Date of birth: {nodo_actual.date_of_birth}','\n')
            print(f'Nationallity: {nodo_actual.nationallity}','\n')
            print(f'Phone: {nodo_actual.phone}','\n')
            print(f'Email: {nodo_actual.email}','\n')
            print(f'Address: {nodo_actual.address}','\n')
            print(f'Career: {nodo_actual.career}','\n')
            self.en_orden(nodo_actual.right)


  # Ejemplo de uso
  arbol = ArbolBinario()

  # Crear instancias de la clase Estudiante
  #estudiante = Student(1, "John Doe", "2000-01-01", 21, "USA", "123-456-7890", "john@example.com", "123 Main St", "Computer Science")
  estudiante = Student(
    student_id_input.value,
    student_name_input.value,
    student_date_of_birth_input.value,
    ages,
    student_nationallity_input.value,
    student_phone_input.value,
    student_email_input.value,
    student_address_input.value,
    student_career_dropdown.value
  )


  # Insertar estudiantes en el árbol
  arbol.insertar(estudiante)


  # Imprimir el árbol en orden por edad
  arbol.en_orden(arbol.root)
  return arbol


register_button_student.on_click(Arbolbinary)
```

```python
class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, student):
        self.root = self._insert(self.root, student)

    def _insert(self, node, student):
        if node is None:
            return student
        if student.age < node.age:
            node.left = self._insert(node.left, student)
        else:
            node.right = self._insert(node.right, student)
        return node

    def pre_order_traversal(self, node):
        if node:
            print(f'ID: {node.id}','\n')
            print(f'Name: {node.name}','\n')
            print(f'Age: {node.age}','\n')
            print(f'Date of birth: {node.date_of_birth}','\n')
            print(f'Nationallity: {node.nationallity}','\n')
            print(f'Phone: {node.phone}','\n')
            print(f'Email: {node.email}','\n')
            print(f'Address: {node.address}','\n')
            print(f'Career: {node.career}','\n')
            self.pre_order_traversal(node.left)
            self.pre_order_traversal(node.right)

# Ejemplo de uso
def arbol_binary():

  #Calcula las fechas para sacar edades
  fecha = str(student_date_of_birth_input.value)
  ages = calcular_edad(fecha)

  estudiante = Student(
    student_id_input.value,
    student_name_input.value,
    student_date_of_birth_input.value,
    ages,
    student_nationallity_input.value,
    student_phone_input.value,
    student_email_input.value,
    student_address_input.value,
    student_career_dropdown.value
  )

  # Crear el árbol e insertar el estudiante
  tree = BinaryTree()
  tree.insert(estudiante)

  # Realizar un recorrido preorden e imprimir los detalles de los estudiantes
  tree.pre_order_traversal(tree.root)
  return tree

register_button_student.on_click(arbol_binary)
```

```python
def facial_recognition():
  count = 1
  # start streaming video from webcam
  video_stream()
  # label for video
  label_html = 'Capturing...'
  bbox = ''

  while True:
    js_reply = video_frame(label_html, bbox)
    if not js_reply:
      break

    # convert JS response to OpenCV Image
    frame = js_to_image(js_reply["img"])

    # Get the face locations from the frame
    face_locations = face_recognition.face_locations(frame)

    # For each face in the frame
    for (top, right, bottom, left) in face_locations:
        # Draw a rectangle around the face
        cv2.rectangle(
            frame,
            (left, top),
            (right, bottom),
            (0, 0, 255),
            2,
        )
        # Get the face encoding of the face
        face_encoding = face_recognition.face_encodings(frame, [(top, right, bottom, left)])[0]

        # Compare the face encoding to the faces in the database
        matches = face_recognition.compare_faces(list(registered_faces.values()), face_encoding)

        # If there is a match
        if True in matches:
          # Get the index of the match
          match_index = matches.index(True)
          # Get the name of the match
          #---------------------------------------------------------------------------------------------------------------
          match_name = list(registered_faces.keys())[match_index]
          cv2.imwrite('/content/drive/MyDrive/framesPath/' +'nombre'+ match_name + str(count) + '.jpg', frame)
          count+=1

        # Ejemplo de uso:
          imagen_persona_path = '/content/drive/MyDrive/framesPath/' +'nombre'+ match_name + str(count) + '.jpg'
          asistencia = Asistencia(match_name, imagen_persona_path)
          lista_asistencia.append(asistencia)
          print(asistencia)

          #--------------------------------------------------------------------------------------------------------------------
          # Draw the name of the match on the frame
          cv2.putText(
              frame,
              match_name,
              (left, top - 10),
              cv2.FONT_HERSHEY_SIMPLEX,
              1,
              (0, 0, 255),
              2,
              cv2.LINE_AA,
          )
          time.sleep(0)
        else:
            # Draw unknown on the frame
            cv2.putText(
                frame,
                "Unknown",
                (left, top - 10),
                cv2.FONT_HERSHEY_SIMPLEX,
                1,
                (0, 0, 255),
                2,
                cv2.LINE_AA,
            )
    # Show image
    plt.imshow(frame)
    plt.show()
```

```python
def bubble_sort():
  with open('LoD_Student.bin', 'rb') as archivo:
    # Carga los datos desde el archivo JSON
    datos = pickle.load(archivo)
    n = len(datos)

    # Recorremos la lista
    for i in range(n):
        # Marcador para saber si se realizó algún intercambio
        intercambio = False

        # Últimos 'i' elementos ya están en su lugar
        for j in range(0, n - i - 1):
            # Comparamos el elemento actual con el siguiente
            if datos[j]["Age"] > datos[j + 1]["Age"]:
                # Intercambiamos los elementos
                datos[j], datos[j + 1] = datos[j + 1], datos[j]
                intercambio = True

        # Si no se realizó ningún intercambio en esta pasada, la lista está ordenada
        if not intercambio:
            break

    i = 1
    # Itera a través de la lista de diccionarios
    html_label1 = widgets.HTML(
    value=(f'<h1 style="color: white; font-size: 25px;">Students:</h1>')
    )
    display(html_label1)
    # Itera a través de la lista de diccionarios
    for element in datos:


        html_label2 = widgets.HTML(
        value=(f'<h1 style="color: white; font-size: 18px;">Students {i}:</h1>')
        )

        id = element['ID']
        name = element['Name']
        dateofbirth = element['Date of Birth']
        age = element['Age']
        nationallity = element['Nationallity']
        phone = element['Phone']
        email = element['Email']
        address = element['Address']
        career = element['Career']
        img = element['Image']
        i += 1

        display(html_label2)
        print(f'Name: {name}','\n')
        print(f'ID: {id}','\n')
        print(f'Date of birth: {dateofbirth}','\n')
        print(f'Age: {age}','\n')
        print(f'Nationallity: {nationallity}','\n')
        print(f'Phone: {phone}','\n')
        print(f'Email: {email}','\n')
        print(f'Address: {address}','\n')
        print(f'Career: {career}','\n')
        print(f'Image: {img}','\n')
    display(back_reader_button)
```

```python
def contar_y_eliminar_nombre(nombre):
  lista = retorna_ListaStudents()
  count = 0  # Inicializamos el contador a 0
  lista_original = lista.copy()  # Creamos una copia de la lista original para evitar modificarla directamente

  for _ in range(lista_original.count(nombre)):
      lista.remove(nombre)  # Eliminamos el nombre de la lista
      count += 1  # Incrementamos el contador en 1 por cada ocurrencia
  return count
```
